<h2>4. Use groupby() to plot the number of "Hamlet" films made each decade</h2>

```python
hamlet_movies_clean = title_hamlet.sort_values(by='year').drop_duplicates(subset=['year'])
hamlet_movies_clean
```

```python
hamlet_movies_clean['Decade'] = (hamlet_movies_clean.year // 10) * 10 
hamlet_movies_clean
```

```python
hamlet_movies_clean.groupby('Decade').count()['year']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
hamlet_movies_clean.groupby('Decade').count()['year'].plot(kind='bar')
```

<h2> 5. How many leading (n=1) roles were available to actors, and how many to actresses, in each year of the 1950s? </h2>

```python
cast[(cast.year >= 1950) & (cast.year <= 1959) & (cast.n == 1)]
```

from the data above I see that 9201 actors and actress were n=1 during those years but there are many actors that played in multiple movies as leading role or even actors that played two leading rols in same movie.
lets remove those who played more then one leading role in the movie.

```python
cast_of_50s = cast[(cast.year >= 1950) & (cast.year <= 1959) & (cast.n == 1)].drop_duplicates(subset=['title','year', 'name'])
cast_of_50s
```

```python
cast_of_50s.groupby('year')['type'].value_counts()
```

this is the answer - in this answer we may count same actor\actress more then once if he playes leading role in more then 1 movie in year. 
if we want to count him once for year - we can use this:

```python
cast_of_50s = cast[(cast.year >= 1950) & (cast.year <= 1959) & (cast.n == 1)].drop_duplicates(subset=['year', 'name'])
cast_of_50s.groupby('year')['type'].value_counts()
```

```python
cast_of_50s.sort_values(by='year')
```

we can see that the secound filter has less rows after the filtering. this is not strange as we filter more data.
